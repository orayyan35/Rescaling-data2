```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python

x = [1, 1, 2, 2, 3, 4, 5, 5]  # Number of rooms in the house
y = [120, 135, 133, 140, 147, 139, 153, 159]
```

```python

def rescaling(x):
    min_X = np.min(x)
    max_X = np.max(x)
    range_X = max_X - min_X
    mean_X = np.mean(x)
    
    def hyp(range_X):
        x_X = [(xi - mean_X) / range_X for xi in x]
        return x_X
    
    x_X = hyp(range_X)
    return x_X
```

```python
print("Original x:", x)
x_rescaled = rescaling(x)
print("Rescaled x:", x_rescaled)
```

```python
print("Original y:", y)
y_rescaled = rescaling(y)
print("Rescaled y:", y_rescaled)
```

```python
# Plotting
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, c='blue', label='Original Data')
plt.xlabel('Number of Rooms')
plt.ylabel('House Price')
plt.title('Original Data')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_rescaled, y_rescaled, c='red', label='Rescaled Data')
plt.xlabel('Rescaled Number of Rooms')
plt.ylabel('Rescaled House Price')
plt.title('Rescaled Data')
plt.legend()

plt.tight_layout()
plt.show()





```
